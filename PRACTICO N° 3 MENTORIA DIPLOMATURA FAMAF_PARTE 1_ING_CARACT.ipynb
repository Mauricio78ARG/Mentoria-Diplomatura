```python
import datetime as dt
```

```python
# VARIABLE FECHA DE STRING A DATETIME##
df['Fecha2'] = pd.to_datetime(df['FECHA'], format='%d/%m/%Y %I:%M:%S %p')
```

```python
df
```

```python
## OBTENEMOS DIA DEL MES EN NOMBRE Y NUMERO##
df['dia_semana']=df['Fecha2'].dt.day_name()
df['dia_numero']= df.Fecha2.dt.strftime('%d')
```

```python
df.head(3)
```

```python
##OBTENEMOS HORA DEL DÍA##
df['hora'] = df['Fecha2'].dt.hour
df.hora.value_counts()
```

```python
## TRANSFORMAMOS LAS ETIQUETAS DE LA VARIABLE CONTRATO EN CATEGORIAS NUMÉRICAS###
from sklearn.preprocessing import StandardScaler, LabelEncoder
df["CONTRATO2"]= LabelEncoder().fit_transform(df['CONTRATO'])
```

```python
df_3=df.drop_duplicates(subset=['CONTRATO2', 'LINEA'])
## CATEGORIAS DE CONTRATOS= 0 (ADULTOS MAYORES), 1 (BEG), 2(BOS), 3(BSC), 4(Discapacidad), 5(Resto), 6(usuarios comunes)
```

```python
## HAY MISSINGS ?##
df.isna().sum()
```

```python
## GENERAMOS DUMMIES QUE INDIQUEN DIA DE LA SEMANA, DIA DEL MES, HORA, SENTIDO Y LINEA##
##df = pd.get_dummies(df, columns=['SENTIDO', 'LINEA', 'dia_semana', 'dia_numero', 'hora'])
##PORQUE SE GENERA UNA BASE QUE NO PUEDE CORRER LA COMPU, SE USA DIA PROMEDIO Y HORA PROMEDIO##
df = pd.get_dummies(df, columns=['SENTIDO', 'LINEA', 'dia_semana',])

```

```python
df.head()
```

```python
y=df.CONTRATO2
```

```python
## ELIMINAMOS COLUMNAS QUE YA NO SIRVEN##
df = df.drop(['FECHA', 'Fecha2', 'CONTRATO'], axis = 1)
```

```python
## OBTENEMOS HORA PROMEDIO Y DIA PROMEDIO POR TARJETA##
df['hora_promedio']= df.groupby('TARJETA')['hora'].transform('mean')
df['dia_promedio']= df.groupby('TARJETA')['dia_numero'].transform('mean')
```

```python
## SUMAMOS LA CANTIDAD DE VIAJES POR DIA DE LA SEMANA##
df['cant_lunes']= df.groupby('TARJETA')['dia_semana_Monday'].transform('sum')
df['cant_martes']= df.groupby('TARJETA')['dia_semana_Tuesday'].transform('sum')                           
df['cant_miercoles']= df.groupby('TARJETA')['dia_semana_Wednesday'].transform('sum') 
df['cant_jueves']= df.groupby('TARJETA')['dia_semana_Thursday'].transform('sum')                                
df['cant_viernes']= df.groupby('TARJETA')['dia_semana_Friday'].transform('sum')                             
df['cant_sabado']= df.groupby('TARJETA')['dia_semana_Saturday'].transform('sum') 
df['cant_domingo']= df.groupby('TARJETA')['dia_semana_Sunday'].transform('sum')                             
                             
```

```python
## CONTAMOS LOS VIAJES DE IDA Y VUELTA POR TARJETA##
df['cant_ida']= df.groupby('TARJETA')['SENTIDO_Ida'].transform('sum')
df['cant_vuelta']= df.groupby('TARJETA')['SENTIDO_Vuelta'].transform('sum')
df
```
